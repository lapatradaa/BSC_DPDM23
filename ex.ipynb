```python
import re
def clean_row(row):
    # Remove parentheses and their contents
    row = re.sub(r'\([^)]*\)', '', row)
    # Remove numbers
    row = re.sub(r'[0-9]', '', row)
    return row.strip()
```

```python
# Apply the clean_row function to each row in the list
cleaned_rows = [clean_row(row) for row in excercise]
```

```python
cleaned_rows_1 = pd.DataFrame({'กิจกรรม': cleaned_rows})
```

```python
cleaned_rows_2 = cleaned_rows_1['กิจกรรม'].str.split(',', expand=True)
```

```python
df_c['กิจกรรม'] = cleaned_rows_2[0]
```

### สร้างฟังก์ชันลบค่า outliers และหาค่า min max q1 q2 q3

```python
def detail_boxreturn(boxreturn):
    q1 = boxreturn['boxes'][0].get_ydata()[0]
    q2 = boxreturn['medians'][0].get_ydata()[0]
    q3 = boxreturn['boxes'][0].get_ydata()[2]
    minv = min(boxreturn['whiskers'][0].get_ydata())
    maxv = max(boxreturn['whiskers'][1].get_ydata())
    print(f'min = {minv}, max = {maxv}, q1 = {q1}, q2 = {q2}, q3 = {q3}')
```

```python
detail_boxreturn(bp_age)
```
